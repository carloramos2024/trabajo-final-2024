## Tarea 1: Marcar todos los sitios de lanzamiento en un mapa

#### Primero, intentemos agregar la ubicación de cada sitio en un mapa usando las coordenadas de latitud y longitud del sitio.

#### El siguiente conjunto de datos con el nombre `spacex_launch_geo.csv` es un conjunto de datos aumentado con latitud y longitud agregadas para cada sitio.

```python
# Download and read the `spacex_launch_geo.csv`
spacex_csv_file = wget.download('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv')
spacex_df=pd.read_csv(spacex_csv_file)
```

#### Ahora, puedes echar un vistazo a cuáles son las coordenadas de cada sitio.

```python
# Seleccione subcolumnas relevantes: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

Las coordenadas anteriores son simplemente números simples que no pueden brindarte ninguna información intuitiva sobre dónde están esos sitios de lanzamiento. Si eres muy bueno en geografía, puedes interpretar esos números directamente en tu mente. Si no, eso también está bien. Visualicemos esas ubicaciones marcándolas en un mapa.

#### Primero debemos crear un objeto `Mapa` en forma de folio, con una ubicación central inicial que será el Centro Espacial Johnson de la NASA en Houston, Texas.

# Tarea 2: Marcar los lanzamientos exitosos/fallidos de cada sitio en el mapa

A continuación, intentemos mejorar el mapa agregando los resultados de lanzamiento de cada sitio y veamos qué sitios tienen tasas de éxito altas.
Recuerde que el marco de datos spacex_df tiene registros de lanzamiento detallados y la columna `class` indica si este lanzamiento fue exitoso o no.

```python
spacex_df.tail(10)
```

A continuación, crearemos marcadores para todos los registros de lanzamiento.
Si un lanzamiento fue exitoso `(class=1)`, entonces usamos un marcador verde y si un lanzamiento falló, usamos un marcador rojo `(class=0)`

Tenga en cuenta que un lanzamiento solo se produce en uno de los cuatro sitios de lanzamiento, lo que significa que muchos registros de lanzamiento tendrán exactamente la misma coordenada. Los grupos de marcadores pueden ser una buena forma de simplificar un mapa que contiene muchos marcadores que tienen la misma coordenada.

#### Primero, creemos un objeto `MarkerCluster`

```python
marker_cluster = MarkerCluster()
```

#### _TODO:_ Crear una nueva columna en el marco de datos `launch_sites` llamada `marker_color` para almacenar los colores de los marcadores según el valor de `class`

```python
# Aplicar una función para verificar el valor de la columna `class`
# Si class=1, el valor de marker_color será verde
# Si class=0, el valor de marker_color será rojo

# Crear la nueva columna 'marker_color' en función de la columna 'class'
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Verificar los primeros valores
spacex_df[['class', 'marker_color']].head()

```

```python
# Función para asignar color al resultado del lanzamiento
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

#### _TODO:_ Para cada resultado de lanzamiento en el marco de datos `spacex_df`, agregue un `folium.Marker` a `marker_cluster`

```python
# Agregar marker_cluster al mapa del sitio actual
site_map.add_child(marker_cluster)

# Para cada fila en el marco de datos spacex_df
# Crear un objeto Marcador con sus coordenadas
# Personalizar la propiedad del icono del Marcador para indicar si este lanzamiento fue exitoso o fallido
for index, record in spacex_df.iterrows():
    # Crear el marcador
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],  # Coordenadas del sitio de lanzamiento
        icon=folium.Icon(
            color='white',  # Fondo del ícono
            icon_color=record['marker_color'],  # Color dinámico según éxito o fallo
            icon='info-sign'  # Ícono representativo
        ),
        popup=f"Site: {record['Launch Site']}<br>Class: {record['class']}"  # Información emergente
    )
    # Agregar el marcador al grupo de marcadores
    marker_cluster.add_child(marker)

# Mostrar el mapa
site_map
```

# TAREA 3: Calcular las distancias entre un sitio de lanzamiento y sus proximidades


#### A continuación, debemos explorar y analizar la proximidad de los sitios de lanzamiento.

#### Primero, agreguemos una `Posición del mouse` en el mapa para obtener las coordenadas del mouse sobre un punto del mapa. De esta manera, mientras explora el mapa, puede encontrar fácilmente las coordenadas de cualquier punto de interés (como el ferrocarril).

```python
# Agregue la posición del mouse para obtener las coordenadas (latitud, longitud) al pasar el mouse sobre el mapa
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Ahora acérquese a un sitio de lanzamiento y explore su proximidad para ver si puede encontrar fácilmente algún ferrocarril, autopista, costa, etc. Mueva el mouse a estos puntos y marque sus coordenadas (mostradas en la parte superior izquierda) para indicar la distancia al sitio de lanzamiento.


#### Puede calcular la distancia entre dos puntos en el mapa basándose en sus valores de "Lat" y "Long" utilizando el siguiente método:

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

#### _TODO:_ Marque un punto en la costa más cercana usando MousePosition y calcule la distancia entre el punto de la costa y el sitio de lanzamiento.

```python
# encuentra las coordenadas de la línea de costa más cercana
# p. ej.: Lat: 28.56367 Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

from math import sin, cos, sqrt, atan2, radians

# Función para calcular la distancia entre dos puntos geográficos
def calculate_distance(lat1, lon1, lat2, lon2):
    # Radio aproximado de la Tierra en kilómetros
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Coordenadas del sitio de lanzamiento (por ejemplo, CCAFS SLC-40)
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Coordenadas de la costa más cercana (encontradas manualmente o con el mouse)
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calcular la distancia entre el sitio de lanzamiento y la costa
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Mostrar el resultado
print(f"La distancia entre el sitio de lanzamiento y la costa más cercana es {distance_coastline:.2f} km.")
```

```python
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    color='blue',
    weight=2.5,
    opacity=1
).add_to(site_map)

# Agregar un marcador en la costa
folium.Marker(
    location=[coastline_lat, coastline_lon],
    icon=folium.Icon(color='blue', icon='info-sign'),
    popup="Coastline"
).add_to(site_map)
```

```python
# Crea un marcador con la distancia a la ciudad, vía férrea, autopista, etc. más cercana.
# Dibuja una línea entre el marcador y el sitio de lanzamiento

from folium import Map, Marker, PolyLine
from folium.plugins import MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Función para calcular la distancia entre dos puntos dados sus valores de latitud y longitud
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Radio aproximado de la Tierra en kilómetros

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Coordenadas del sitio de lanzamiento (ejemplo: CCAFS SLC-40)
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Coordenadas de la ciudad, ferrocarril, autopista, etc. más cercana obtenida con MousePosition (ejemplo)
nearest_location_lat = 28.401897  # Coordenadas de ejemplo
nearest_location_lon = -80.601806

# Calcular la distancia entre el sitio de lanzamiento y la ubicación más cercana
distance_to_nearest = calculate_distance(launch_site_lat, launch_site_lon, nearest_location_lat, nearest_location_lon)

# Crear el mapa
site_map = Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Agregar un marcador en el sitio de lanzamiento
Marker([launch_site_lat, launch_site_lon], popup="Sitio de lanzamiento").add_to(site_map)

# Agregar un marcador en la ubicación más cercana
distance_marker = Marker(
    [nearest_location_lat, nearest_location_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_to_nearest),
    ),
    popup="Ubicación más cercana"
)
distance_marker.add_to(site_map)

# Dibujar una línea entre el sitio de lanzamiento y la ubicación más cercana
line = PolyLine(
    locations=[
        [launch_site_lat, launch_site_lon],
        [nearest_location_lat, nearest_location_lon]
    ],
    weight=2,
    color='blue'
)
site_map.add_child(line)

# Mostrar el mapa
site_map
```
